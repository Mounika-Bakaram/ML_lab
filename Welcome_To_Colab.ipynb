```
#1. Write a python program to compute Central Tendency Measures: Mean, Median, Mode Measure of Dispersion: Variance, Standard Deviation

import statistics

# Real-time data: Student exam scores
exam_scores = [78, 85, 90, 88, 76, 95, 89, 92, 85, 87]

# Central Tendency Measures
mean_value = statistics.mean(exam_scores)          # Mean
median_value = statistics.median(exam_scores)      # Median
mode_value = statistics.mode(exam_scores)          # Mode

# Measures of Dispersion
variance_value = statistics.variance(exam_scores)  # Variance
std_dev_value = statistics.stdev(exam_scores)      # Standard Deviation

# Display the results
print(f"Mean (Average Score): {mean_value}")
print(f"Median (Middle Score): {median_value}")
print(f"Mode (Most Frequent Score): {mode_value}")
print(f"Variance (Score Spread): {variance_value}")
print(f"Standard Deviation (Score Deviation): {std_dev_value}")
```
